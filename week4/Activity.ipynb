## With Stemming:

```python
import nltk
from nltk.stem import LancasterStemmer
stemmerLan = LancasterStemmer()
```

```python
text3 = []
for i in text1:
    text3.append(stemmerLan.stem(i))

text3
```

```python
text4 = []
for i in text2:
    text4.append(stemmerLan.stem(i))

text4
```

```python
vect1 = CountVectorizer(binary=True)
```

```python
vect1.fit(text3)
```

```python
vocab1 = vect1.vocabulary_
vocab1
```

```python
transformed = vect1.transform(text4).toarray()
print(len(transformed))
```

```python
similar=[] # will store index corresponding to similar vocab
for i in range(len(transformed)):
    for j in range(len(transformed[i])):
        if(transformed[i][j]>0):
            similar.append((i,j))

print(len(similar))
print(similar)
```
